```python
import pandas as pd
import glob
import os
from risk_indexes import *

file_paths = glob.glob("../data/dataset/*.csv")
results = []
```

```python
for file in file_paths:

    dataframe = pd.read_csv(file)

    stock_name = os.path.basename(file).replace(".csv", "")
    
    if 'Close' not in dataframe.columns:
        print(f"Skipping {file}. It does not have a 'Close' column!")
        continue
    
    returns = dataframe["Close"].pct_change().dropna()
    
    if len(returns) == 0:
        print(f"{stock_name} stock has no returns")
        continue
    
    results.append(
        {
            "Stock": stock_name,
            "Annual_Return": annualized_return(returns),
            "Volatility": annualized_volatility(returns),
            "Sharpe": sharpe_ratio(returns),
            "Sortino": sortino_ratio(returns),
            "VaR_95": value_at_risk(returns),
            "CVaR_95": conditional_var(returns),
            "Max_Drawdown": max_drawdown(dataframe["Close"]),
        }
    )
```

```python
risk_df = pd.DataFrame(results)
risk_df.to_csv("../outputs/risk_metrics.csv", index=False)
```
